```python
from flask import Flask
import subprocess
import threading

app = Flask(__name__)

def run_lms():
    while True:
        subprocess.run(["lms", "server", "start", "--port", "1234"])

threading.Thread(target=run_lms, daemon=True).start()

@app.route("/")
def home():
    return "LM Studio Server Running"

app.run(host="0.0.0.0", port=5000)
```
